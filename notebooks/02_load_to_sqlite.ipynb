```python
import pandas as pd
import sqlite3
import os
```

```python
# Create a path for our database file
db_path = '../data/flights.db'

# delete the file if it exists from a previous run
if os.path.exists(db_path):
    os.remove(db_path)
    print(f"Removed existing database at {db_path}")

# Connect (creates the file if it doesn't exist)
conn = sqlite3.connect(db_path)
print(f"Database created at {db_path}")
```

```python
# Load the full CSV in chunks of 100,000 rows
csv_path = '../data/raw/flights_sample_3m.csv'
chunk_size = 100_000  # process 100k rows at a time

# Counter
total_rows = 0

print("Loading data into SQLite... this will take a few minutes.")

for chunk_num, chunk in enumerate(pd.read_csv(csv_path, chunksize=chunk_size)):
    # Write each chunk to the 'flights' table; append after the first
    if_exists = 'replace' if chunk_num == 0 else 'append'
    chunk.to_sql('flights', conn, if_exists=if_exists, index=False)
    
    total_rows += len(chunk)
    print(f"  Chunk {chunk_num + 1}: loaded {total_rows:,} rows so far...")

print(f"\nDone! Total rows loaded: {total_rows:,}")
```

```python
# verification load worked — count rows in the table created
cursor = conn.cursor()
cursor.execute("SELECT COUNT(*) FROM flights")
row_count = cursor.fetchone()[0]
print(f"Rows in 'flights' table: {row_count:,}")
```

```python
# close connection
conn.close()
print("Connection closed.")
```
